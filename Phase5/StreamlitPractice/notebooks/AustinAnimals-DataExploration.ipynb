```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
df = pd.read_csv('data/Intakes_2021_withlocationdetails.csv')
```

```python
df.head()
```

```python
df.info()
```

### Animal Type Exploration

```python
df['Animal Type'].value_counts(normalize=True)
```

```python
types = df['Animal Type'].value_counts()
```

```python
types = types.reset_index().rename(columns={'index':'Type', 'Animal Type':'Count'})
```

```python
types
```

```python
for an_type in types.index:
    print(f"{types['Type'][an_type]}: {types['Count'][an_type]}")
```

```python
fig = px.pie(
    types,
    names='Type',
    values='Count',
)
fig.show()
```

### Location Analysis

```python
df.head()
```

```python
df['Found Location'].value_counts()
```

```python
has_loc = df[~df['found_lat'].isna()]
```

```python
has_loc.info()
```

```python
has_loc['Date'] = pd.to_datetime(has_loc['DateTime']).dt.date
```

```python
has_loc['Date']
```

```python
import datetime
```

```python
datetime.date(2021,5,1)
```

```python
has_loc.loc[has_loc['Date'] > datetime.date(2021,5,1)]
```

```python
has_loc['Date'].min()
```

```python
sub_df = has_loc.groupby(['Found Location', 'found_lat', 'found_lon'])['Animal ID'].count().reset_index()
```

```python
sub_df = sub_df.rename(columns={'Animal ID':'count'})
```

```python
sub_df.head()
```

```python
fig = px.scatter_mapbox(sub_df, lat="found_lat", lon="found_lon", 
                        hover_name="Found Location", size='count', color='count', zoom=9)

fig.update_layout(
    mapbox_style="open-street-map",
    margin={"r":0,"t":0,"l":0,"b":0},
    mapbox=dict(
        center={'lat': 30.2672, 'lon': -97.7431},
    )
)

fig.show()
```
